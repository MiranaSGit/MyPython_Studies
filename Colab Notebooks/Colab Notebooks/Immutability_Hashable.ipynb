**Now Let's focus on Lists**

**By comparing Lists and Tuples after some operations and/or applying for some methods, we can clearly figure out if their id's can change or not.**

**Now, we will try to expand our list and tuple using the += operator. This operation will work for both data types. Letâ€™s see what will happen.**

```python
list_values = [1, 2, 3]
set_values = (1, 2, 3)
print(list_values)
print(set_values)
print(id(list_values))
print(id(set_values))
print()

list_values += [4, 5, 6]
set_values += (4, 5, 6)
print(list_values)
print(set_values)
print(id(list_values))
print(id(set_values))
```

**We can see that the list identity is not changed, while the tuple identity is changed. This means that we have expanded (modified) our list, but created (assigned) a completely new tuple. Lists are more memory efficient than tuples.**

```python
my_list = [10, 20, 30]
print(my_list)
print(id(my_list))
print()
my_list[0] = 40
print(my_list)
print(id(my_list))
```
